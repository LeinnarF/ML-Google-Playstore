```python
import pandas as pd
import numpy as np

raw_df = pd.read_csv('dataset/googleplaystore.csv')
raw_df
```

```python
# Get the number of rows with invalid ratings (5 < Rating and NaN)
greater_than_five = raw_df[raw_df['Rating'] > 5]
less_than_1 = raw_df[raw_df['Rating'] < 1]
is_nan = raw_df[raw_df['Rating'].isnull()]
is_string = raw_df[raw_df['Rating'].apply(lambda x: isinstance(x, str))]

print(f"Number of rows with rating greater than 5: {len(greater_than_five)}")
print(f"Number of rows with rating less than 1: {len(less_than_1)}")
print(f"Number of rows with NaN rating: {len(is_nan)}")
print(f"Number of rows with string rating: {len(is_string)}")
```

```python
# Get frequency of each category
category_counts = raw_df['Category'].value_counts()
category_counts
```

```python
genre_counts = raw_df['Genres'].unique()
for genre in genre_counts:
    print(genre)
```

```python
price_counts = raw_df['Price'].value_counts()
print(price_counts)
```
